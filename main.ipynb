```python

import pandas as pd
import numpy as np
import ast

# Hente film data fra csv fil
def get_movie_data():
    movie_data = pd.read_csv("tmdb_5000_movies.csv")
    return movie_data

movie_data = get_movie_data()

id_col = movie_data['id']
title_col = movie_data['title']
genres_col = movie_data['genres']
rating_col = movie_data['vote_average']
evaluation_col = movie_data['vote_count']


def parse_genres(genres_str):
    genres_list = ast.literal_eval(genres_str)
    genres = [genre['name'] for genre in genres_list]
    return ', '.join(genres)

genres_parsed = genres_col.apply(parse_genres)

print("ID:")
print(id_col.head())
print("\nTitle:")
print(title_col.head())
print("\nGenres:")
print(genres_parsed.head())
print("\nRating:")
print(rating_col.head())
print("\nEvaluation:")
print(evaluation_col.head())



id = id_col
title = title_col
genre = genres_parsed
rating = rating_col
evaluation = evaluation_col

# aksserer data basert p√• index
print(id[1])
```
